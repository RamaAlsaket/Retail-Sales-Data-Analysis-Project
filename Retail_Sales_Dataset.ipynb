# Retail Sales Dataset

**Overview**

This dataset captures retail sales transactions for a fictional or real-world store. It’s structured to help analyze sales trends, customer behavior, and product performance.

**Columns:**

1.   Transaction ID
2.   Date
3.   Customer ID
4.   Gender
5.   Age
6.   Product Category
7.   Quantity
8.   Price per Unit
9.   Total Amount



**Key Points**

Data Type Variety: Includes numeric data (quantity, price), categorical data (product), and date/time (order date).

**Potential Insights:**

1. Total sales per product, month, or region
2. Most popular products
3. Revenue trends over time
4. Customer behavior patterns (high-value orders, repeat purchases)


Why it’s interesting? & Goal

- You can easily compute metrics like total sales, average order value, monthly trends, and the most needed product

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Product Category', y='Total Amount', data=data)
plt.xticks(rotation=45)
plt.title("Total Amount by Product")
plt.show()
```

Electronics likely has the highest median "Total Amount", suggesting people spend more on electronics compared to Beauty or Clothing.
Beauty might have the lowest median, indicating lower typical spending.

```python
monthly_sales = data.groupby('Month')['Total Amount'].sum()

plt.figure(figsize=(8,4))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Total Sales per Month")
plt.show()
```

```python
monthly_product_sales = data.groupby(['Month', 'Product Category'])['Total Amount'].sum().reset_index()
```

```python
custom_colors = {
    "Electronics": "#1f77b4",
    "Clothing": "#ff7f0e",
    "Furniture": "#2ca02c",
    "Toys": "#d62728"
}

fig = px.line(
    monthly_product_sales,
    x="Month",
    y="Total Amount",
    color="Product Category",
    markers=True,
    title="Monthly Product Sales by Category",
    color_discrete_map=custom_colors
)

fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Total Sales Amount",
    legend_title="Product Category",
    template="plotly_white"
)

fig.show()
```
